```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Categorical Data

### a. Countplot

```python
sns.countplot(x=df['Survived'])
df['Survived'].value_counts()
```

### b. Piechart

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

# 2. Numerical Data

### a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5) #do check what happens if you remove the bins=5 argument
```

### b. Distplot

```python
sns.distplot(df['Age'])
#this is called a kernel density fxn/ probability density fxn
#also tells us if data has skewness
```

### c. Boxplot

```python
sns.boxplot(df['Age']) #do check what a boxplot is and how it displays data
```

```python
df['Age'].min()
```
